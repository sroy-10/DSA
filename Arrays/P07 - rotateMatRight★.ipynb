## Problem Statement

You have been given a matrix ‘MAT’ of size 'N' * 'M' (where 'N' and 'M' denote the number of rows and columns respectively) and a positive integer ‘K’. Your task is to rotate the matrix to the right 'K' times.

**Note:**\
Right rotation on a matrix is shifting each column to the right side (the last column moves to the first column) and 'K' times means performing this rotation 'K' times.

**Example:**\
For 'K' = 1 and the given 'MAT':

1 2 3\
4 5 6\
7 8 9

Output after rotating 'MAT' one time is:

3 1 2 \
6 4 5\
9 7 8


**Constraints:**\
1 <= T <= 10\
1 <= N <= 200\
1 <= M <= 200\
0 <= K <= 10^9\
1 <= MAT[i][j] <= 10^5

Where 'MAT[i][j]' denotes the element in the 'i'th row and 'j'th column of the matrix.

It is guaranteed that sum of 'N' * 'M' over all the test cases does not exceed 10 ^ 5.

**Time limit:** 1 sec

## Algorithm

To solve this problem with optimal time and space complexity, you should realize that rotating the matrix 'K' times to the right is equivalent to shifting the columns by 'K' positions. If 'K' is larger than 'M', you can take 'K' modulo 'M' to minimize the number of rotations, as rotating the matrix 'M' times will result in the original matrix.

**Here's a step-by-step guide on how to solve the problem:**
1. For each test case:
    1. Read the matrix 'MAT' of size 'N' * 'M'.
    1. Read the value of 'K'.

1. Since rotating the matrix 'M' times results in the original matrix, calculate the effective number of rotations as 'K' = 'K' % 'M'.

1. Perform the rotation operation:
    1. For each row, shift the elements to the right by 'K' positions.
    1. To do this efficiently, you can use an auxiliary array or perform the operation in place with some clever indexing.

1. Output the rotated matrix in the specified row-wise format.

## Implementation

```python
def rotateMatRight(mat, n, m, k):
    # Write your code here.
    k = k % m
    if k == 0:
        return mat
    else:
        for i in range(n):
            mat[i] = mat[i][-k:] + mat[i][:-k]
        return mat
```
